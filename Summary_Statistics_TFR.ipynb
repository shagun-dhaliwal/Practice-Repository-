# Creating dataframe with only countries in Africa

Create a new dataframe which only has countries in Africa.


```
TFR_ss =dataTFR[dataTFR['Region'] == 'Sub-Saharan Africa']
TFR_ss.head()
#But still missing 6 African countries from region "Middle East & North Africa"
```

```
#TFR_NA = dataTFR[dataTFR['Country Name']==['Egypt, Arab Rep.','Libya', 'Tunisia' ,'Algeria','Morocco','Djibouti']]
countries = ['Egypt, Arab Rep.','Libya', 'Tunisia' ,'Algeria','Morocco','Djibouti']

new = dataTFR['Country Name'].isin(countries)
dataTFR[new]
```

```
#Concat Sub-Saharan and North African countries into one dataframe
TFR_Africa = pd.concat([TFR_ss,dataTFR[new]], ignore_index=True)
TFR_Africa
```

```
TFR_Africa.to_csv('TFR_Africa_Cleaned.csv')
```
